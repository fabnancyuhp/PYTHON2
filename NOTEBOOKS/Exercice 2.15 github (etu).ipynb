## Correction 
* Dans cet exercice on doit utiliser les tables pour trouver les quantiles théoriques. 
* Les quantiles théoriques seront arrondies à 2 chiffres aprés la virgule.

1. La série x est de taille $n=10$. On a le tableau suivant:
    $$\begin{array}{|c|cccccccccc|}

    x_{(i)} &  -0.93 & -0.91& -0.57& -0.21 &  0.09 & 0.34 & 0.65 & 0.7 & 1.22 & 1.89\\
    \hline
    F_{i} & \frac{1}{10} & \frac{2}{10} & \frac{3}{10} & \frac{4}{10} & \frac{5}{10} & \frac{6}{10} & \frac{7}{10} & \frac{8}{10} & \frac{9}{10} & 1\\

    \hline
    \end{array}
    $$
    

On va page 12 des tables:
* page 12 à gauche, ligne 0.1 $x_{(1)}^{*} = -1.2816\simeq -1.28$
* page 12 à gauche, ligne 0.2 $x_{(2)}^{*} = -0.8416\simeq -0.84$
* page 12 à gauche, ligne 0.3 $x_{(3)}^{*} = -0.5244\simeq -0.52$
* page 12 à gauche, ligne 0.4 $x_{(4)}^{*} = -0.5244\simeq -0.25$
* page 12 à droite, ligne 0.5 $x_{(5)}^{*} = 0.00$
* page 12 à droite, ligne 0.6 $x_{(6)}^{*} = 0.2533\simeq 0.25$
* page 12 à droite, ligne 0.7 $x_{(7)}^{*} = 0.5244\simeq 0.52$
* page 12 à droite, ligne 0.8 $x_{(8)}^{*} = 0.8416\simeq 0.84$
* page 12 à droite, ligne 0.9 $x_{(9)}^{*} = 1.2816\simeq 1.28$

Les points $(x_{(i)},x_{i}^{*})$ sont alignés sur la bissectrice. Il y a donc adéquation avec la loi $\mathcal{N}(0,0)$

2. La série est de taille $n=10$
$$\begin{array}{|c|cccccccccc|}
    \hline
    x_{(i)} &  1.67 & 2.25 & 3.06 & 3.07 & 3.11 & 3.12 & 3.44 & 4.11 & 4.44 & 4.56\\
    \hline
    F_{i} & \frac{1}{10} & \frac{2}{10} & \frac{3}{10} & \frac{4}{10} & \frac{5}{10} & \frac{6}{10} & \frac{7}{10} & \frac{8}{10} & \frac{9}{10} & 1\\
    \hline
    \phi^{-1}(i/10)=\frac{x_{(i)}^{*}-3}{1} & -1.28 & -0.84& -0.52 & -0.25 &  0. & 0.25 & 0.52 & 0.84 & 1.28 & \\
    \hline
    x_{i}^{*} = \phi^{-1}(i/10)+3 & 1.72 & 2.16 & 2.48 & 2.75 & 3. & 3.25 & 3.52 & 3.84 & 4.28 &\\
    \hline
    \end{array}
    $$

Si la série de nombre x provient de la loi $\mathcal{N}(3,1)$ alors $\frac{x-3}{1}$ provient d'une loi $\mathcal{N}(0,1)$. On désigne par $\phi$ la fonction de répartition de la loi $\mathcal{N}(0,1)$. les $x_{i}^{*}$ sont les quantiles théoriques de $\mathcal{N}(3,1)$. Alors $\frac{x_{i}^{*}-3}{1}$ sont les quantiles théoriques de $\mathcal{N}(0,1)$. On a donc 
$\frac{x_{i}^{*}-3}{1}=\phi^{-1}(i/10)$ et $x_{i}^{*} = \phi^{-1}(i/10)+3$.

Les points $(x_{(i)},x_{i}^{*})$ sont alignés sur la bissectrice. Il y a donc adéquation avec la loi $\mathcal{N}(3,1)$

Si la série de nombre x provient de la loi $\mathcal{N}(-1,2)$ alors $\frac{x+1}{2}$ provient d'une loi $\mathcal{N}(0,1)$. On désigne par $\phi$ la fonction de répartition de la loi $\mathcal{N}(0,1)$. les $x_{i}^{*}$ sont les quantiles théoriques de $\mathcal{N}(-1,2)$.
